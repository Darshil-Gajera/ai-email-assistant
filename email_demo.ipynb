# Load Free Model
*   Use Flan-T5 (good for writing, replying, summarizing emails).




```python
from transformers import pipeline

# Load model
email_writer = pipeline("text2text-generation", model="google/flan-t5-base")

# Test it
prompt = "Write a formal email to HR requesting 2 days leave due to personal reasons."
result = email_writer(prompt, max_length=200)
print(result[0]['generated_text'])
```

*   Write New Email

```python
task = input("Enter task (write/reply/summarize): ")
tone = input("Tone (formal/friendly): ")
content = input("Enter email prompt or text: ")

full_prompt = f"Task: {task}. Tone: {tone}. Content: {content}"
print(email_writer(full_prompt, max_length=200)[0]['generated_text'])
```
